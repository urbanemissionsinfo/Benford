```python
def calc_euclidean(obs_freq):
    dist = np.sqrt(np.sum((obs_freq - expected_freq)**2))
    return dist
```

```python
def bootstrap_digit_frequencies(data, confidence_level=0.95):
    n = len(data)
    unique_digits = np.arange(10)
    
    # Bootstrap samples
    bootstrap_samples = np.random.choice(data, size=(n, n), replace=True)
    
    # Calculate frequencies for each bootstrap sample
    bootstrap_freqs = np.array([np.bincount(sample, minlength=10) / n for sample in bootstrap_samples])
    
    # Percentiles for the confidence intervals
    lower_percentile = (1 - confidence_level) / 2 * 100
    upper_percentile = (1 + confidence_level) / 2 * 100
    
    # Calculate bootstrap confidence intervals for each digit
    freqs = np.bincount(data, minlength=10) / n
    freqs = [n*prop for prop in freqs]
    ci_lower = np.percentile(bootstrap_freqs, lower_percentile, axis=0)
    ci_lower = [n*prop for prop in ci_lower]
    ci_upper = np.percentile(bootstrap_freqs, upper_percentile, axis=0)
    ci_upper = [n*prop for prop in ci_upper]

    
    
    return freqs[1:], ci_lower[1:], ci_upper[1:]
```

```python
import warnings
warnings.filterwarnings("ignore")

cities = []
station_nums = []
eds = []
years = []
for city in tqdm(df['City'].unique()):
    df_city = df[df['City']==city]
    df_city['leading_digit'] = df_city['Index Value'].astype(str).str[0:1].astype(int)
    for year in [2018, 2019, 2020, 2021, 2022, 2023, 2024]:
        ed = np.nan
        mean_station_num = np.nan
        df_city_year = df_city[df_city['year']==year]
        
        if df_city_year.shape[0] < 300:
            continue
            
        expected_counts = [i*df_city_year.shape[0] for i in expected_freq]
        mean_station_num = df_city_year['No. Stations'].astype(int).mean()
        observed_freq = np.array(pd.merge(check,
                        df_city_year.sort_values('leading_digit').leading_digit.value_counts(normalize=True),
                        on='leading_digit',
                        how='left').fillna(0)['proportion'])
        ed = calc_euclidean(observed_freq)
        cities.append(city)
        station_nums.append(mean_station_num)
        eds.append(ed)
        years.append(year)
```

# K-s
**Null Hypothesis:** Both samples belong to the same distribution.

Generally, same distribution is not the same "type" of distribution. Even two normal distributions would fail - if their means/stds are different.

But when we do mean normalisation, then even samples from normal distributions with different means and stds would still hold KS test. 

In this case, KS test will reject only when the "type" is different -- normal vs uniform etc.

```python
from scipy.stats import kstest, gaussian_kde

def normalise_array(arr):
    mean = arr.mean()
    std = arr.std()
    max = arr.max()
    min = arr.min()
    arr_normalised = arr - mean
    arr_normalised = arr_normalised/(max-min)
    return arr_normalised
```

```python
np.random.seed(441)
sample1 = np.random.normal(100, 10, 1000)
#sample2 = np.random.uniform(1, 10, 1000) #np.random.normal(1, 100, 1000)
sample2 =np.random.normal(10, 10, 1000)

#sample1 = np.random.lognormal(0, 0.1, 1000)
#sample2 = np.random.lognormal(0.3, 0.3, 1000)
#sample1 = normalise_array(sample1)
#sample2 = normalise_array(sample2)

ks_statistic, p_value = kstest(sample1, sample2)
# Print the results
print("KS Statistic:", ks_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis. Both samples are from different distributions")
else:
    print("Fail to reject the null hypothesis. Both samples may not be from different distributions")

plt.hist(sample1, density=True)
plt.hist(sample2, density=True, alpha=0.6)
plt.show()
```

```python
from scipy import integrate
import warnings
warnings.filterwarnings("ignore")
city = 'Udaipur'
df_city = df[df['City']==city]

fig, axes = plt.subplots(2, 5, figsize=(28,14))
axes = axes.flatten()

count=0
for year in [2019, 2020, 2021, 2022, 2023, 2019, 2020, 2021, 2022, 2023]:
    if count <5:
        base = 2018
    else:
        base = year-1
    #base = 2018# year - 1
    sample1 = df_city[df_city['year']==base]['Index Value']
    sample2 = df_city[df_city['year']==year]['Index Value']
    sample1 = normalise_array(sample1)
    sample2 = normalise_array(sample2)
    
    #axes[year-2019].hist(sample1, density=True)
    #axes[year-2019].hist(sample2, density=True, alpha=0.6)
    kde1 = sns.kdeplot(sample1, ax=axes[count])
    x1, y1 = kde1.get_lines()[0].get_data()
    kde2 = sns.kdeplot(sample2, ax=axes[count])
    x2, y2 = kde2.get_lines()[1].get_data()

    # Ensure both KDEs are evaluated over the same x-axis range
    common_x = np.linspace(min(x1.min(), x2.min()), max(x1.max(), x2.max()), 1000)

    # Interpolate KDE values to the common x-axis range
    y1_interp = np.interp(common_x, x1, y1)
    y2_interp = np.interp(common_x, x2, y2)

    # Total area under KDE1 and KDE2
    total_area = 2 # KDE = PDF -- total area under one KDE has to be 1. 
    # Calculate the overlap area by integrating the minimum of both KDE curves
    overlap_area = integrate.simps(np.minimum(y1_interp, y2_interp), common_x)

    non_overlap_area = 100*(total_area - 2*overlap_area)/total_area
    
    ks_statistic, p_value = kstest(sample1, sample2)
    # Interpret the results
    alpha = 0.05  # Significance level
    if p_value < alpha:
        KS_interpret = 'Reject Ho'
    else:
        KS_interpret = 'Fail to Reject Ho'

    axes[count].annotate('Non-overlap area:',
                             xy =(0.05, 3.4),
                             fontsize=18)
    axes[count].annotate(str(round(non_overlap_area))+'%',
                             xy =(0.3, 3.1),
                             fontsize=30)
    axes[count].annotate('K-S Test: \n' + KS_interpret,
                             xy =(-0.7, 3.2),
                             fontsize=18)
    
    # Plot the KDEs with overlapping regions shaded
    axes[count].fill_between(common_x, np.minimum(y1_interp, y2_interp), color='purple', alpha=0.4, label='Overlap Area')

    
    axes[count].set_title(str(base)+'-'+str(year), fontsize=24)
    axes[count].set_ylim(0, 3.6)#max(y1.max(), y2.max())+0.4)
    axes[count].set_xlim(-0.8, 1)#max(y1.max(), y2.max())+0.4)
    axes[count].legend(labels=[str(base),str(year)], loc='lower right',prop={'size': 20})
    axes[count].set_xlabel('Normalised AQI value', fontsize=18)
    axes[count].set_ylabel('Density', fontsize=20)
    axes[count].tick_params(axis='both', which='major', labelsize=16)
    count = count+1
plt.suptitle(city, fontsize=30)
fig.text(0.175, -0.02, 'Null Hypothesis (Ho): Two samples are from same distribution', ha='center', fontsize=20)
plt.tight_layout()
#plt.tight_layout(rect=[0, 0.3, 1, 0.97])  # Adjust layout so the annotation isn't cut off

#plt.savefig(root + '/plots/anomalies/' + city+'_yearlycompare.jpg')
plt.savefig(root + '/plots/anomalies/' + city+'_PDFS.jpg', bbox_inches='tight')
plt.show()
```

```python
from scipy import integrate

np.random.seed(441)
sample1 = np.random.normal(100, 10, 1000)
#sample2 = np.random.uniform(70, 130, 1000) #np.random.normal(1, 100, 1000)
sample2 =np.random.normal(100, 10, 1000)
sample1 = normalise_array(sample1)
sample2 = normalise_array(sample2)

kde1 = sns.kdeplot(sample1, )
x1, y1 = kde1.get_lines()[0].get_data()
kde2 = sns.kdeplot(sample2, )
x2, y2 = kde2.get_lines()[1].get_data()

# Ensure both KDEs are evaluated over the same x-axis range
common_x = np.linspace(min(x1.min(), x2.min()), max(x1.max(), x2.max()), 1000)

# Interpolate KDE values to the common x-axis range
y1_interp = np.interp(common_x, x1, y1)
y2_interp = np.interp(common_x, x2, y2)

# Total area under KDE1 and KDE2
total_area = 2 # KDE = PDF -- total area under one KDE has to be 1. 
# Calculate the overlap area by integrating the minimum of both KDE curves
overlap_area = integrate.simps(np.minimum(y1_interp, y2_interp), common_x)

non_overlap_area = 100*(total_area - 2*overlap_area)/total_area

ks_statistic, p_value = kstest(sample1, sample2)
# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    KS_interpret = 'Reject H0 \nDifferent distribution'
else:
    KS_interpret = 'Fail to reject H0 \n(Same distribution)'

plt.annotate('Non-overlap area(%) \n '+str(round(non_overlap_area,2)),
                         xy =(max(x1.max(), x2.max())-1.3, max(y1.max(), y2.max())-0.7),
                         fontsize=14)
plt.annotate('2 sample KS Test: \n' + KS_interpret,
                         xy =(max(x1.max(), x2.max())-1.3, max(y1.max(), y2.max())-0.32),
                         fontsize=14)

# Plot the KDEs with overlapping regions shaded
plt.fill_between(common_x, np.minimum(y1_interp, y2_interp), color='purple', alpha=0.4, label='Overlap Area')
plt.xlabel('Normalised values')
plt.legend(['Sample1 KDE', 'Sample2 KDE'], loc='upper right')
plt.tight_layout()
plt.savefig(root + '/plots/' +'nonoverlap.jpg')
```

```python
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

areas = []
for i in tqdm(range(0,10000)):
    sample1 = np.random.normal(0, 1, 365)
    sample2 =np.random.normal(0, 1, 365)
    sample1 = normalise_array(sample1)
    sample2 = normalise_array(sample2)
    
    kde1 = sns.kdeplot(sample1, )
    x1, y1 = kde1.get_lines()[0].get_data()
    kde2 = sns.kdeplot(sample2, )
    x2, y2 = kde2.get_lines()[1].get_data()
    
    # Ensure both KDEs are evaluated over the same x-axis range
    common_x = np.linspace(min(x1.min(), x2.min()), max(x1.max(), x2.max()), 1000)
    
    # Interpolate KDE values to the common x-axis range
    y1_interp = np.interp(common_x, x1, y1)
    y2_interp = np.interp(common_x, x2, y2)
    
    # Total area under KDE1 and KDE2
    total_area = 2 # KDE = PDF -- total area under one KDE has to be 1. 
    # Calculate the overlap area by integrating the minimum of both KDE curves
    overlap_area = integrate.simps(np.minimum(y1_interp, y2_interp), common_x)
    
    non_overlap_area = 100*(total_area - 2*overlap_area)/total_area
    plt.close()

    areas.append(non_overlap_area)
```
